```python
 #Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import os
from pprint import pprint
```

```python
author_url="http://127.0.0.1:5000/api/v1.0/merged_author"
author_response = requests.get(author_url).json()
pprint(author_response)
```

```python
import json
with open('author.json', 'w') as json_file:
    json.dump(author_response, json_file)
```
